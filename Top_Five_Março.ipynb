```
data['LEAD'] = data['LEAD'].str.replace(',','.')
data['LEAD'] = data['LEAD'].astype(float)
data['TITULO'] = data['TITULO'].str.replace(',','.')
data['TITULO'] = data['TITULO'].astype(float).round(2)
data['LEAD'] = data['LEAD'].astype(float).round(2)
```

##Melhores Notas De Acordo Com a quantidade de avaliações, tiulo, lead, quantidade de sim e não

```
avaliacoes = (data['SIM'] + data['NÃO'])
data['Avaliações'] = pd.Series(data=avaliacoes.round(2))
```

```
data.sort_values(['Avaliações', 'TITULO', 'LEAD', 'SIM', 'NÃO'], ascending=[False, False, False, False, True])
```

##Maiores Notas De Acordo Com o Titulo

##Maiores Notas De Acordo Com o LEAD

```
data.sort_values(['LEAD'], ascending=[False])
```

##Média Geral Titulo e Lead

```
media = (data['TITULO'] + data['LEAD']) / 2
```

```
data['Média'] = pd.Series(data=media.round(2))
```

##Maiores Notas De Acordo com a Média

```
data_media = data.sort_values(['Média'], ascending=[False])
```
